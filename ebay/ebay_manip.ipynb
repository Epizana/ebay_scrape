```python
catpltly = ebay1[['brand','category','bids','price']]
catpltly = catpltly.loc[(catpltly['bids'] != 0) & (catpltly['bids'] < 50)]
catpltly
```

```python
catpltly = ebay1[['brand','category','bids','price']]
#catpltly = catpltly.groupby(['category','brand']).count()[['bids']].sort_values(by = 'bids', ascending=False)
cpltly = catpltly.loc[catpltly['brand'].isin(['Hasbro','Mattel','Hot Toys','Unbranded','Kenner','Bandai','Marvel','McFarlane Toys','Star Wars','DC','Funko','Mezco','Takara TOMY'])]#.reset_index()
cpltly = cpltly.loc[cpltly['bids'] != 0]
cpltly = cpltly.loc[~cpltly['category'].isin(['Mixed Lots','Animals & Dinosaurs','Fantasy','Music','Anime & Manga','Other Action Figures'])]
cpltly = cpltly.groupby(['category','brand']).count()[['bids']].sort_values(by = 'bids', ascending=False).reset_index()
cpltly
```

```python
import plotly.express as px
fig1 = px.box(cpltly, x="bids", y="category",points="all", orientation = 'h',
              hover_data=['brand'],
              height=600,width = 880,
              title = 'Bid Variability by Category: Top Brands')

fig1.update_layout(plot_bgcolor='rgb(233,233,233)')
fig1.show()
```
