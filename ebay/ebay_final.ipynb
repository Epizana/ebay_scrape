```python
ebay2 = ebay1[['brand','category','franchise','condition','price','bids','quantity_sold','title','format']]
```

```python
brand_group = ebay2.groupby('brand')
cat_group = ebay2.groupby('category')
cond_group = ebay2.groupby('condition')
```

```python
cond_group_size = pd.DataFrame(cond_group.size())
cond_group_size
```

```python
condition_price = cond_group.agg('sum')[['price']]
total_sales = int(condition_price.sum())
#condition_price
```

```python
brand_by_sales = brand_group.agg('sum')[['price']].sort_values('price',ascending = False).head(20)
top_brands = brand_by_sales.index.tolist()
brand_by_sales
```

```python
cats_by_sales = cat_group.agg('sum')[['price']].sort_values('price',ascending = False).head(5)
top_cats = cats_by_sales.index.tolist()
cats_by_sales
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
brand_by_sales.sort_values(by = 'price', ascending = True).plot.barh(color='b')
plt.xlabel('Total Sales')
plt.ylabel('Brand')
```

```python
ebaychart = ebay1[['brand','category','condition','bids','price','title','format']]
```

```python
ebaysns = ebaychart.loc[ebaychart['brand'].isin(top_brands)].groupby(['brand','condition']).sum()[['price']].sort_values(by = 'price', ascending=False).reset_index()
#ebaysns

plt.rcParams['figure.figsize'] = [12, 8]
sns.set(style = 'whitegrid')
ax = sns.barplot(x = 'price', y = 'brand', hue = 'condition', data = ebaysns)
plt.xlabel('Total Sales')
plt.ylabel('Brand')
#plt.title('Total Sales by top 20 Brands by Item Condition', fontsize=15)
```

```python
facetchart = ebaychart.loc[(ebaychart['brand'].isin(['Hasbro','Unbranded','Mattel','NECA'])) & (ebaychart['category'].isin(top_cats))]
facetchart = facetchart.groupby(['brand','category','condition','format']).mean()[['price']].reset_index()
facetchart.replace({'Comic Book Heroes':'CBH', 'Military & Adventure': 'M&A', 'Sports':'S','TV, Movie & Video Games':'TGMVG','Transformers & Robots':'T'}, inplace=True)
#facetchart
```

```python
fig = px.bar(facetchart, x="format", y="price", color="condition", barmode="group", facet_row="brand", facet_col="category",
             labels = {'brand':'','price':'Average Price','category':''},
             category_orders={"category": top_cats, "brand":['Hasbro','Unbranded','NECA','Mattel']},
             height=600,width = 830,
             title='Average Selling Price by Category, Condition, and Format')
fig.show()
```

```python
hotchart = ebaychart.loc[(ebaychart['brand']=='Hot Toys') & (ebaychart['category'].isin(top_cats))]
hotchart = hotchart.groupby(['brand','category','condition','format']).mean()[['price']].reset_index()
hotchart.replace({'Comic Book Heroes':'CBH', 'Military & Adventure': 'M&A', 'Sports':'S','TV, Movie & Video Games':'TGMVG','Transformers & Robots':'T'}, inplace=True)
#hotchart
```

```python
fig = px.bar(hotchart, x="format", y="price", color="condition", barmode="group", facet_row="brand", facet_col="category",
             labels = {'brand':'','price':'Average Price','category':''},
             category_orders={"category": top_cats, "brand":'Hot Toys'},
             height=600,width = 830,
             title='Hot Toys Average Selling Price by Category, Condition, and Format')
fig.show()
```
